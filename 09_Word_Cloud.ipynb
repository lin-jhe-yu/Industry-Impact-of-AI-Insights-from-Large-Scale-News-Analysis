## Word Cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd # Ensure pandas is imported if not already in context for df_ai_sentiment

# Prepare text data for each sentiment category
positive_text = ' '.join(df_ai_sentiment[df_ai_sentiment['pred_sentiment'] == 'positive']['clean_text'].dropna())
neutral_text = ' '.join(df_ai_sentiment[df_ai_sentiment['pred_sentiment'] == 'neutral']['clean_text'].dropna())
negative_text = ' '.join(df_ai_sentiment[df_ai_sentiment['pred_sentiment'] == 'negative']['clean_text'].dropna())

# Diagnostic: Check text lengths
print(f"Length of positive_text: {len(positive_text)} characters")
print(f"Length of neutral_text: {len(neutral_text)} characters")
print(f"Length of negative_text: {len(negative_text)} characters")
```

```python
# Define common stopwords that might not be useful in word clouds
stopwords = set(WordCloud().stopwords)
# Add any domain-specific or overly common words if needed
# stopwords.update(['ai', 'artificial intelligence', 'new', 'report', 'company'])

# Function to generate and display a word cloud
def plot_word_cloud(text, title, color):
    if not text:
        print(f"Warning: No text available for '{title}'. Skipping word cloud generation.")
        return
    try:
        wordcloud = WordCloud(width=800, height=400, background_color='white', colormap=color, stopwords=stopwords).generate(text)
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title(title, fontsize=16, weight='bold', pad=15)
        plt.show()
    except Exception as e:
        print(f"Error generating word cloud for '{title}': {e}")
```

```python
plot_word_cloud(positive_text, 'Positive Sentiment Word Cloud', 'Greens')
```

```python
plot_word_cloud(neutral_text, 'Neutral Sentiment Word Cloud', 'Blues')
```

```python
plot_word_cloud(negative_text, 'Negative Sentiment Word Cloud', 'Reds')
```

## Word Cloud title (Industry)

```python
entity_to_industry_mapping = validated_entities_df.set_index('entity')['industry'].to_dict()
print(entity_to_industry_mapping)
```

```python
all_industries = merged_df_with_impact['industries_in_title'].explode()
top_10_industries = all_industries.value_counts().head(11)
print("Top 10 most frequent industries:")
print(top_10_industries)
```

```python
selected_industries_for_wordcloud = top_10_industries.index.to_list()
print("Defined selected_industries_for_wordcloud:")
print(selected_industries_for_wordcloud)
```

```python
def identify_industries_in_title(title, entity_to_industry_mapping):
    if not isinstance(title, str):
        return []
    found_industries = set()
    for entity, industry in entity_to_industry_mapping.items():
        if pd.notna(industry) and entity.lower() in title.lower():
            found_industries.add(industry)
    return list(found_industries)

merged_df_with_impact['industries_in_title'] = merged_df_with_impact['title'].apply(lambda x: identify_industries_in_title(x, entity_to_industry_mapping))

print(merged_df_with_impact[['title', 'industries_in_title']].head())

sentiment_colors = {
    'positive': 'Greens',
    'neutral': 'Blues',
    'negative': 'Reds'
}

for industry in selected_industries_for_wordcloud:
    print(f"\nGenerating word clouds for {industry} industry...")
    for sentiment in ['positive', 'neutral', 'negative']:
        sentiment_industry_titles = merged_df_with_impact[
            merged_df_with_impact['industries_in_title'].apply(lambda x: industry in x if isinstance(x, list) else False) &
            (merged_df_with_impact['pred_sentiment'] == sentiment)
        ]['title'].dropna()

        if not sentiment_industry_titles.empty:
            combined_titles = ' '.join(sentiment_industry_titles)
            plot_word_cloud(
                combined_titles,
                f'{sentiment.capitalize()} Sentiment Word Cloud for {industry} Industry',
                sentiment_colors[sentiment]
            )
        else:
            print(f"No {sentiment} titles found for industry: {industry}")
```
