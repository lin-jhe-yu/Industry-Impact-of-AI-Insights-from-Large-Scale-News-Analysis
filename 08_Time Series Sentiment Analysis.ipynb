# Task

3.  **Overall Monthly Sentiment Analysis**: Calculate the average sentiment for all articles per month.
4.  **Company-Specific Monthly Sentiment Analysis**: Analyze average monthly sentiment for articles mentioning specific companies (Microsoft, FCC, Samsung, Amazon, IBM) using the `entity_to_industry_mapping.csv` file.
5.  **Topic-Specific Monthly Sentiment Analysis**: Analyze average monthly sentiment for articles belonging to specific topics (1, 2, 3, 4, 5).
6.  **Industry-Specific Monthly Sentiment Analysis**: Analyze average monthly sentiment for articles related to specific industries (Technology, Media, Finance, Retail/E-commerce) by extracting entities from `clean_text` and mapping them to industries using `entity_to_industry_mapping.csv`.

8.  **Summary**: Provide a summary of the key findings from the time series sentiment analysis across all categories.

I will begin by preparing the data by converting the 'date' column and mapping the `pred_sentiment` to numerical values. Then, I will calculate the overall monthly sentiment.

## Date Preprocessing

### Subtask:
Convert the 'date' column in 'merged_df_with_impact' to datetime objects and extract a 'year_month' column for monthly aggregation. This step will prepare the data for time-series analysis using the existing `pred_sentiment` column.


**Reasoning**:
First, I'll convert the 'date' column to datetime objects to enable date-based operations. Then, I'll extract the year and month from the 'date' column and store them in a new 'year_month' column, which is essential for monthly aggregation.



```python
merged_df_with_impact['date'] = pd.to_datetime(merged_df_with_impact['date'])
merged_df_with_impact['year_month'] = merged_df_with_impact['date'].dt.strftime('%Y-%m')
print("Date column converted to datetime and 'year_month' created.")
merged_df_with_impact[['date', 'year_month']].head()
```

## Sentiment Quantification

### Subtask:
Convert the categorical `pred_sentiment` column ('positive', 'neutral', 'negative') into a numerical format (e.g., 1, 0, -1) for quantitative analysis.

#### Instructions
1. Define a mapping for sentiment: 'positive' to 1, 'neutral' to 0, and 'negative' to -1.
2. Apply this mapping to the `pred_sentiment` column to create a new numerical sentiment column, for example, 'numerical_sentiment'.

**Reasoning**:
I will define a mapping for the sentiment categories and then apply this mapping to the `pred_sentiment` column to create a new numerical sentiment column named `numerical_sentiment`.



```python
sentiment_mapping = {'positive': 1, 'neutral': 0, 'negative': -1}
merged_df_with_impact['numerical_sentiment'] = merged_df_with_impact['pred_sentiment'].map(sentiment_mapping)
print("Sentiment categories mapped to numerical values.")
merged_df_with_impact[['pred_sentiment', 'numerical_sentiment']].head()
```

```python
overall_monthly_sentiment = merged_df_with_impact.groupby('year_month')['numerical_sentiment'].mean().reset_index()
print("Overall monthly sentiment calculated.")
overall_monthly_sentiment.head()
```

## Company-Specific Monthly Sentiment Analysis

### Subtask:
Analyze average monthly sentiment for articles mentioning specific companies (Microsoft, FCC, Samsung, Amazon, IBM) using the `entity_to_industry_mapping` DataFrame.

#### Instructions:
1. Define a list of target companies: `['microsoft', 'fcc', 'samsung', 'amazon', 'ibm']`.
2. Filter `merged_df_with_impact` to include only articles that mention any of these target companies in their `clean_text`.
3. Group the filtered data by `year_month` and the company mentioned.
4. Calculate the mean of `numerical_sentiment` for each group.
5. Store the results in a new DataFrame, for example, `company_monthly_sentiment`.

**Reasoning**:
I will define the list of target companies, then filter the `merged_df_with_impact` to find articles mentioning these companies in their `clean_text`. Finally, I will group the filtered data by `year_month` and the mentioned company to calculate the average numerical sentiment for each.



```python
import numpy as np

target_companies = ['microsoft', 'fcc', 'samsung', 'amazon', 'ibm']

# Function to check if any target company is mentioned in clean_text
def get_mentioned_company(text, companies):
    if isinstance(text, str):
        for company in companies:
            if company in text.lower():
                return company
    return np.nan

# Apply the function to create a new column 'mentioned_company'
merged_df_with_impact['mentioned_company'] = merged_df_with_impact['clean_text'].apply(lambda x: get_mentioned_company(x, target_companies))

# Filter out rows where no target company was mentioned
company_specific_df = merged_df_with_impact.dropna(subset=['mentioned_company'])

# Calculate monthly sentiment for each company
company_monthly_sentiment = company_specific_df.groupby(['year_month', 'mentioned_company'])['numerical_sentiment'].mean().reset_index()

print("Company-specific monthly sentiment calculated.")
company_monthly_sentiment.head()
```

## Topic-Specific Monthly Sentiment Analysis

### Subtask:
Analyze average monthly sentiment for articles belonging to specific topics (1, 2, 3, 4, 5).

#### Instructions:
1. Define a list of target topics: `[1, 2, 3, 4, 5]`.
2. Filter `merged_df_with_impact` to include only articles that belong to any of these target topics in their `topic` column.
3. Group the filtered data by `year_month` and `topic`.
4. Calculate the mean of `numerical_sentiment` for each group.
5. Store the results in a new DataFrame, for example, `topic_monthly_sentiment`.

**Reasoning**:
I will define the list of target topics, then filter the `merged_df_with_impact` to include articles from these topics. Finally, I will group the filtered data by `year_month` and `topic` to calculate the average numerical sentiment for each.



```python
target_topics = [2, 3, 5, 6, 8]

# Filter out articles belonging to target topics
topic_specific_df = merged_df_with_impact[merged_df_with_impact['topic'].isin(target_topics)]

# Calculate monthly sentiment for each topic
topic_monthly_sentiment = topic_specific_df.groupby(['year_month', 'topic'])['numerical_sentiment'].mean().reset_index()

print("Topic-specific monthly sentiment calculated.")
topic_monthly_sentiment.head()
```

## Industry-Specific Monthly Sentiment Analysis

### Subtask:
Analyze average monthly sentiment for articles related to specific industries (Technology, Media, Finance, Retail/E-commerce) by extracting entities from `clean_text` and mapping them to industries using `entity_to_industry_mapping.csv`.

#### Instructions:
1. Define a list of target industries: `['Technology', 'Media', 'Finance', 'Retail/E-commerce']`.
2. Iterate through each row of `merged_df_with_impact` to identify articles that mention entities belonging to the target industries.
3. For each article, extract entities from `clean_text` and check if they map to any of the target industries using `entity_to_industry_mapping`.
4. Group the data by `year_month` and the identified industry.
5. Calculate the mean of `numerical_sentiment` for each group.
6. Store the results in a new DataFrame, for example, `industry_monthly_sentiment`.

**Reasoning**:
I will define the list of target industries and create a mapping of entities to industries. Then, I will apply a function to the `clean_text` column to identify all relevant industries for each article, handling multiple industry mentions by exploding the DataFrame. Finally, I will group the data by `year_month` and the identified industry to calculate the average numerical sentiment.



```python
target_industries = ['Technology / AI Platforms', 'Government / Policy / Regulation', 'Finance / FinTech', 'Healthcare & Biotech', 'Retail & E-commerce', 'Automotive / Mobility']

# Create entity-to-industry dictionary for quick lookup, filtering for target industries
filtered_entity_industry_map = entity_to_industry_mapping[
    entity_to_industry_mapping['industry'].isin(target_industries)
]
entity_industry_dict = {
    str(entity).lower(): industry
    for entity, industry in filtered_entity_industry_map[['entity', 'industry']].values
}

# Function to find all relevant industries mentioned in text
def get_mentioned_industries(text, entity_map, target_industries_list):
    if not isinstance(text, str):
        return []

    found_industries = set() # Use a set to store unique industries
    text_lower = text.lower()

    for entity_name_lower, industry in entity_map.items():
        if entity_name_lower in text_lower and industry in target_industries_list:
            found_industries.add(industry)

    return list(found_industries)

# Apply the function to create a new column 'related_industries'
merged_df_with_impact['related_industries'] = merged_df_with_impact['clean_text'].apply(
    lambda x: get_mentioned_industries(x, entity_industry_dict, target_industries)
)

# Explode the DataFrame to have one row per (article, industry) pair, then filter for rows with identified industries
industry_specific_df = merged_df_with_impact[
    merged_df_with_impact['related_industries'].apply(len) > 0
].explode('related_industries')

# Calculate monthly sentiment for each industry
industry_monthly_sentiment = industry_specific_df.groupby(
    ['year_month', 'related_industries']
)['numerical_sentiment'].mean().reset_index()

# Rename the 'related_industries' column to 'industry' for clarity
industry_monthly_sentiment.rename(columns={'related_industries': 'industry'}, inplace=True)

print("Industry-specific monthly sentiment calculated.")
industry_monthly_sentiment.head()
```

## Visualization

### Subtask:
Generate plots to visualize all the calculated sentiment trends.

#### Instructions:
1. Plot the `overall_monthly_sentiment` to visualize the overall sentiment trend over time.
2. Plot the `company_monthly_sentiment` to visualize sentiment trends for each target company.
3. Plot the `topic_monthly_sentiment` to visualize sentiment trends for each target topic.
4. Plot the `industry_monthly_sentiment` to visualize sentiment trends for each target industry.
5. Ensure all plots have appropriate titles, axis labels, and legends for clarity.
6. Use `matplotlib.pyplot` and `seaborn` for plotting.

**Reasoning**:
I will begin by importing the necessary libraries for plotting, `matplotlib.pyplot` and `seaborn`, then proceed to create the first visualization for the overall monthly sentiment.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure year_month is treated as a datetime for proper plotting
overall_monthly_sentiment['year_month'] = pd.to_datetime(overall_monthly_sentiment['year_month'])

plt.figure(figsize=(12, 6))
sns.lineplot(data=overall_monthly_sentiment, x='year_month', y='numerical_sentiment')
plt.title('Overall Monthly Sentiment Trend')
plt.xlabel('Month')
plt.ylabel('Average Numerical Sentiment')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print("Overall monthly sentiment trend plot displayed.")
```

```python
company_monthly_sentiment['year_month'] = pd.to_datetime(company_monthly_sentiment['year_month'])

plt.figure(figsize=(14, 7))
sns.lineplot(data=company_monthly_sentiment, x='year_month', y='numerical_sentiment', hue='mentioned_company', marker='o')
plt.title('Company-Specific Monthly Sentiment Trends')
plt.xlabel('Month')
plt.ylabel('Average Numerical Sentiment')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(title='Company')
plt.tight_layout()
plt.show()
print("Company-specific monthly sentiment trends plot displayed.")
```

## Visualize Sentiment Trends


```python
topic_monthly_sentiment['year_month'] = pd.to_datetime(topic_monthly_sentiment['year_month'])

# Merge with topic_mapping to get topic names
topic_monthly_sentiment_named = topic_monthly_sentiment.merge(
    topic_mapping[['topic_id', 'topic_name']],
    left_on='topic',
    right_on='topic_id',
    how='left'
)

# Filter out topic 1
topic_monthly_sentiment_named = topic_monthly_sentiment_named[topic_monthly_sentiment_named['topic'] != 1]

plt.figure(figsize=(14, 7))
sns.lineplot(data=topic_monthly_sentiment_named, x='year_month', y='numerical_sentiment', hue='topic_name', marker='o')
plt.title('Topic-Specific Monthly Sentiment Trends')
plt.xlabel('Month')
plt.ylabel('Average Numerical Sentiment')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(title='Topic Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
print("Topic-specific monthly sentiment trends plot displayed with topic names and excluding topic 1.")
```
