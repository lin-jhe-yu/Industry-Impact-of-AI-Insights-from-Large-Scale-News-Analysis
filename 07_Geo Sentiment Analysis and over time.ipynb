## Geographic Sentiment Analysis

```python
import numpy as np

# Recreate df_with_sentiment to ensure 'locations' column is in its original format from 'df'
df_with_sentiment = pd.merge(df, merged_df_with_impact[['news_id', 'pred_sentiment']], on='news_id', how='left')

print("Data type of 'locations' column:")
print(df_with_sentiment['locations'].dtype)

print("Value counts of types in 'locations' column:")
print(df_with_sentiment['locations'].apply(type).value_counts())

print("First 10 non-null values from 'locations' column:")
print(df_with_sentiment['locations'].dropna().head(10))
```

```python
locations_exploded_df = df_with_sentiment[df_with_sentiment['locations'].apply(len) > 0].explode('locations')

# Display the first few rows to verify
print("First 5 rows of locations_exploded_df:")
locations_exploded_df.head()
```

**Reasoning**:
Now that the `locations` column is correctly exploded, I need to map each individual location to its corresponding country using `geotext` and `pycountry` and then aggregate these country mentions. This is the next logical step towards performing sentiment analysis by country.



```python
from geotext import GeoText
import pycountry

def get_country_from_location(location_name):
    # Prioritize exact match with pycountry
    try:
        country = pycountry.countries.lookup(location_name)
        return country.name.lower()
    except LookupError:
        pass

    # Fallback to GeoText for broader recognition
    places = GeoText(location_name)
    if places.countries:
        # GeoText returns ISO alpha-2 codes or full names. Convert to full name.
        # Attempt to get full name if ISO code is returned
        iso_codes = places.countries
        for code in iso_codes:
            try:
                country = pycountry.countries.get(alpha_2=code) # Try with alpha_2
                return country.name.lower()
            except KeyError:
                try:
                    country = pycountry.countries.get(name=code) # Try with full name
                    return country.name.lower()
                except KeyError:
                    pass
        if iso_codes: # If GeoText found something but pycountry couldn't resolve
            return iso_codes[0].lower() # Return the first one as is
    return 'unknown'

# Apply the function to the 'locations' column of the exploded DataFrame
locations_exploded_df['country'] = locations_exploded_df['locations'].apply(get_country_from_location)

print("First 5 rows of locations_exploded_df with country:")
display(locations_exploded_df.head())
```

**Reasoning**:
With the `locations_exploded_df` now containing a 'country' column, the next step is to group this DataFrame by 'country' and 'pred_sentiment' to count the occurrences of each sentiment type for each country. This will allow for the calculation of sentiment distribution later.



```python
sentiment_counts = locations_exploded_df.groupby(['country', 'pred_sentiment']).size().unstack(fill_value=0)
sentiment_counts['total'] = sentiment_counts.sum(axis=1)

print("First 5 rows of sentiment_counts:")
display(sentiment_counts.head())
```

**Reasoning**:
To fulfill the subtask, I need to calculate the percentage distribution of each sentiment type (negative, neutral, positive) for every country. This will be achieved by dividing the sentiment counts by the total mentions for each country.



```python
sentiment_by_country = sentiment_counts.copy()
sentiment_by_country['negative_percentage'] = (sentiment_by_country['negative'] / sentiment_by_country['total'] * 100).round(2)
sentiment_by_country['neutral_percentage'] = (sentiment_by_country['neutral'] / sentiment_by_country['total'] * 100).round(2)
sentiment_by_country['positive_percentage'] = (sentiment_by_country['positive'] / sentiment_by_country['total'] * 100).round(2)

# Sort by total mentions to see the most active countries first
sentiment_by_country = sentiment_by_country.sort_values(by='total', ascending=False)

print("First 5 rows of sentiment_by_country (percentage distribution):")
display(sentiment_by_country.head())
```

## Identify Top Countries by Sentiment Score

```python
sentiment_by_country['net_sentiment_score'] = sentiment_by_country['positive_percentage'] - sentiment_by_country['negative_percentage']

# Sort by net sentiment score in descending order for most positive
top_positive_sentiment = sentiment_by_country.sort_values(by='net_sentiment_score', ascending=False)
print("\nTop 10 Countries with Most Positive Net Sentiment:")
display(top_positive_sentiment[['total', 'net_sentiment_score']].head(10))

# Sort by net sentiment score in ascending order for most negative
top_negative_sentiment = sentiment_by_country.sort_values(by='net_sentiment_score', ascending=True)
print("\nTop 10 Countries with Most Negative Net Sentiment:")
display(top_negative_sentiment[['total', 'net_sentiment_score']].head(10))
```

## Temporal Geographic Sentiment Trends

### Subtask:
Analyze how sentiment towards AI-related news changes over time for specific key countries, identifying trends or events that might have influenced public opinion.


**Reasoning**:
The first step is to convert the 'date' column to datetime objects to enable temporal analysis, handling any conversion errors by coercing invalid dates to NaT.



```python
locations_exploded_df['date'] = pd.to_datetime(locations_exploded_df['date'], errors='coerce')
print("Date column converted to datetime:")
display(locations_exploded_df[['date', 'pred_sentiment', 'country']].head())
```

**Reasoning**:
Now that the 'date' column is in datetime format, I will define the key countries, filter the DataFrame, group the data by date, country, and sentiment, unstack the sentiment to create separate columns, calculate total mentions, and then compute the percentage distribution of each sentiment type over time for these selected countries.



```python
selected_countries = ['united states', 'india', 'china']

# Filter for selected countries and drop rows with NaT dates or unknown countries
temporal_df = locations_exploded_df[
    (locations_exploded_df['country'].isin(selected_countries)) &
    (locations_exploded_df['date'].notna())
].copy()

# Extract year and month for monthly aggregation
temporal_df['year_month'] = temporal_df['date'].dt.to_period('M')

# Group by year_month, country, and sentiment, then count occurrences
temporal_sentiment_counts = temporal_df.groupby(['year_month', 'country', 'pred_sentiment']).size().unstack(fill_value=0)

# Calculate total mentions for each month and country
temporal_sentiment_counts['total'] = temporal_sentiment_counts.sum(axis=1)

# Calculate percentages
temporal_sentiment_counts['negative_percentage'] = (temporal_sentiment_counts['negative'] / temporal_sentiment_counts['total'] * 100).round(2)
temporal_sentiment_counts['neutral_percentage'] = (temporal_sentiment_counts['neutral'] / temporal_sentiment_counts['total'] * 100).round(2)
temporal_sentiment_counts['positive_percentage'] = (temporal_sentiment_counts['positive'] / temporal_sentiment_counts['total'] * 100).round(2)

print("First 5 rows of temporal sentiment percentages for selected countries (monthly):")
display(temporal_sentiment_counts.head())
```

**Reasoning**:
To visualize how sentiment changes over time for the selected countries, I will create line plots for negative, neutral, and positive percentages. This will help identify trends and potential influential events.



```python
import matplotlib.pyplot as plt

# Reset index to make 'year_month' and 'country' accessible as columns for plotting
temporal_sentiment_plot_df = temporal_sentiment_counts.reset_index()

# Convert 'year_month' to datetime for plotting
temporal_sentiment_plot_df['year_month'] = temporal_sentiment_plot_df['year_month'].dt.to_timestamp()

# Calculate net sentiment for each time point
temporal_sentiment_plot_df['net_sentiment_percentage'] = temporal_sentiment_plot_df['positive_percentage'] - temporal_sentiment_plot_df['negative_percentage']

# Plotting net sentiment trends for each country
fig, ax = plt.subplots(figsize=(15, 6))
fig.suptitle('Monthly Net Sentiment Trends for Key Countries', fontsize=16)

# Define a list of distinct colors for the countries
country_colors = {'united states': 'blue', 'india': 'red', 'china': 'green'}

for country in selected_countries:
    country_df = temporal_sentiment_plot_df[temporal_sentiment_plot_df['country'] == country]
    # Assign a unique color to each country
    ax.plot(country_df['year_month'], country_df['net_sentiment_percentage'], label=country.capitalize(), color=country_colors.get(country, 'black'), alpha=0.7)

ax.set_title('Monthly Net Sentiment (Positive % - Negative %) Over Time')
ax.set_ylabel('Net Sentiment Percentage')
ax.legend()
ax.grid(True)

plt.xlabel('Date')
plt.tight_layout(rect=[0, 0.03, 1, 0.96])
plt.show()
```

```python
import re

def clean_title(title):
    # Convert to lowercase
    title = title.lower()
    # Remove punctuation
    title = re.sub(r'[^"\w\s"]', '', title)
    # Remove common trailing news source names (example, extend as needed)
    common_suffixes = ['abc news', 'wwlp', 'cna', 'wtoptv', 'youreriecom', 'rochesterfirst', 'yourbasin', 'bnn bloomberg', 'news 4 buffalo', 'wgn radio 720  chicagos very own', 'wntza', 'krqe news 13', 'kelolandcom', 'wspa 7news', 'wavycom', 'wtmj', 'fox40', 'cbs 42', 'cnn', 'bbc', 'reuters', 'the new york times']
    for suffix in common_suffixes:
        if title.endswith(suffix):
            title = title[:-len(suffix)].strip()
    # Consolidate multiple spaces and remove leading/trailing spaces
    title = re.sub(r'\s+', ' ', title).strip()
    return title

if not china_2022_sentiment.empty:
    dip_month_info = china_2022_sentiment.loc[china_2022_sentiment['net_sentiment_percentage'].idxmin()]
    dip_year_month = dip_month_info['year_month']

    dip_month_period = pd.Period(dip_year_month, freq='M')

    # Filter for negative sentiment during the dip period
    negative_news_for_dip = locations_exploded_df[
        (locations_exploded_df['country'] == 'china') &
        (locations_exploded_df['date'].dt.to_period('M') == dip_month_period) &
        (locations_exploded_df['pred_sentiment'] == 'negative')
    ]

    negative_dip_news_ids = negative_news_for_dip['news_id'].unique()

    # Retrieve titles from the original df_with_sentiment using these news_ids
    raw_negative_dip_titles = df_with_sentiment[df_with_sentiment['news_id'].isin(negative_dip_news_ids)]['title'].unique()

    # Apply cleaning function and then get unique titles
    cleaned_negative_dip_titles = [clean_title(title) for title in raw_negative_dip_titles]
    unique_cleaned_negative_dip_titles = pd.Series(cleaned_negative_dip_titles).unique()

    print(f"Top 20 CLEANED news titles with NEGATIVE sentiment related to China during the {dip_year_month.strftime('%Y-%m')} sentiment dip:")
    if len(unique_cleaned_negative_dip_titles) > 0:
        for i, title in enumerate(unique_cleaned_negative_dip_titles[:20]):
            print(f"{i+1}. {title}")
    else:
        print("No cleaned news titles with negative sentiment found for this period.")
else:
    print("No sentiment data available for China in 2022.")
```
